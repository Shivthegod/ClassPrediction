```python
import networkx as nx
from networkx.drawing.nx_pydot import graphviz_layout
import pydot
```

```python
G = nx.DiGraph()

with open('prereqs.txt') as fp:
    lines = fp.readlines()

classes = set()
for line in lines:
    split = line.split(" ")
    classes.add(split[0].replace("\n", ""))
    classes.add(split[1].replace("\n", ""))

G.add_nodes_from(classes)

for line in lines:
    split = line.split(" ")
    G.add_edge(split[0].replace("\n", ""), split[1].replace("\n", ""))
```

```python
nx.draw(G, pos=graphviz_layout(G, prog="twopi"), with_labels=True, node_size=1500)
```
